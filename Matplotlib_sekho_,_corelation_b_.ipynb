```
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```
x = df["longitude"]
y = df["latitude"]
```

```
a = x-x.mean() 
b=  y-y.mean()
```

```
ab = a*b
a2 = a**2
b2 = b**2
```

```
a2summ = a2.sum()
b2summ = b2.sum()
absumm = ab.sum()
```

```
corre =  absumm/np.sqrt(a2summ*b2summ)
```
